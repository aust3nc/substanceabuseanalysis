filtering to show useage only; renaming

```python
useage = data.filter(items = ['alcohol-use', 'marijuana-use', 'cocaine-use', 'crack-use', 'heroin-use', 'hallucinogen-use', 'inhalant-use', 'pain-releiver-use', 'oxycontin-use', 'tranquilizer-use', 'stimulant-use', 'meth-use', 'sedative-use'])
useage = useage.rename(columns = {'alcohol-use': 'alcohol', 'marijuana-use': 'marijuana', 'cocaine-use': 'cocaine', 'crack-use': 'crack', 'heroin-use': 'heroin', 'hallucinogen-use': 'hallucinogen', 'inhalant-use': 'inhalant', 'pain-releiver-use': 'pain-reliever', 'oxycontin-use': 'oxycontin', 'tranquilizer-use': 'tranquilizer', 'stimulant-use': 'stimulant', 'meth-use': 'meth', 'sedative-use': 'sedative'})
useage.head()
```

let's do some light visual analysis; I'll start with a horizontal bar summating drug useage for all ages

```python
keys = list(useage.columns)
for key in keys:
    values = [sum(useage[key]) for key in keys]

for keys, values in zip(keys, values):
    plt.barh(keys, values)
```

let's focus on alcohol, our most usual suspect...

```python
alcohol = data.filter(items = ['age', 'alcohol-use'])
age = list(alcohol['age'])
drinking = list(alcohol['alcohol-use'])

plt.pie(drinking, labels=age, autopct='%1.1f%%')
plt.title('Alchoholism By Age Group')
plt.axis('equal')
plt.show()
```

it appears wisdom comes with age, or lack of--the majority of alcoholics fall in the center of our range. next, we'll explore interactively...

```python
def makebarh(df):
    print(list(df.columns))
    title = str(input('Input the name of the substance you would like to analyse' ))
    target = df[title]
    plt.pie(target, labels = age, autopct='%1.1f%%')
    plt.axis('equal')
    plt.title(title)
    plt.show()

makebarh(useage)
```
